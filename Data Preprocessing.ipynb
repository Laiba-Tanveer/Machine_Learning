# Outlier Detection and Removal

__Checking the structure of the dataset__

```python
data.describe()
```

```python
0.25-1.5*0.5
```

```python
0.75 + 1.5 * 0.5
```

__Identify numeric columns__

```python
numeric_cols = data.select_dtypes(include=[np.number])
```

__Calculate IQR__

```python
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
```

__Filter out outliers        # 0.25-1.5*0.5 = -0.5                      #  0.75 + 1.5 * 0.5 = 1.5__

```python
data_cleaned = data[~((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)]
```

__Visualizing the data before and after outlier removal__

```python
plt.figure(figsize=(20, 6))
```

__Box plot before removing outliers__

```python
plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1) 
numeric_cols.boxplot()
plt.title("Before Outlier Removal")
# plt.xticks(rotation=45)  
plt.tight_layout()   
```

__Box plot after removing outliers__

```python
plt.figure(figsize=(20, 6))

# Box plot after removing outliers
plt.subplot(1, 2, 2)
data_cleaned.select_dtypes(include=[np.number]).boxplot()
plt.title("After Outlier Removal")

plt.tight_layout()
plt.show()
```

# Data Transformation

## Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

__Load the dataset__

```python
data = pd.read_csv("Student_Performance.csv")
```

__Separate numeric and non-numeric columns__

```python
numeric_cols = data.select_dtypes(include=[np.number])
non_numeric_cols = data.select_dtypes(exclude=[np.number])
```

__Apply MinMaxScaler to the numeric columns__

```python
scaler = MinMaxScaler()
scaled_numeric_data = scaler.fit_transform(numeric_cols)
```

__Convert the scaled numeric data back to a DataFrame__

```python
scaled_numeric_df = pd.DataFrame(scaled_numeric_data, columns=numeric_cols.columns)
```

__Combine the scaled numeric columns with the non-numeric columns__

```python
scaled_data = pd.concat([scaled_numeric_df, non_numeric_cols.reset_index(drop=True)], axis=1)
```

__Check the shape of the scaled data__

```python
print(scaled_data.shape)
print()
print('*' * 60)
scaled_data.head()
```
